# Limpieza y valores nulos

Por desgracia, en la mayoría de conjuntos de datos reales faltan datos. Algunos motivos pueden ser:

No se sabe el valor
Se introdujo una nueva columna y por lo tanto no se conocen los valores de dicha columna para datos anteriores
No sabe/no contesta
No se consideró relevante ...
La falta de valores en una tabla supondrá un grave problema en el momento en que queramos construir un modelo predictivo, ya que la mayoría de los modelos no pueden realizar predicciones si no se conocen todos los valores de todas las columnas en cada fila de nuestra tabla.

Empecemos contando cuántos valores faltan en cada columna de nuestro fichero

```python
df_ardilla.isna().sum()
```

## Analizar los valores nulos

Comprobamos que en la tabla numeros, en la tercera columnas contiene la fecha de cuando se vieron, se ha interpretado como numero en vez de fecha, ademas los valores nulos, representan un problema a la hora de predecir modelos predictivos, pues no los admiten

```python
df_ardilla['location'].value_counts(dropna=False)
```

Obervamos que las ardillas no conocemos esos ''_64 casos_'' de valores nulos, la mayoria de los avistamientos son en el **suelo** (Ground Plane) y el resto sobre los **arboles** (Above Ground), con lo cual esos casos nulos los contemplaremos sonre el **suelo**. Veámos cómo reemplazar los valores nulos de esta columna por "Ground Plane".

```python
df_ardilla['location']=df_ardilla['location'].fillna('Ground Plane')
df_ardilla['location'].value_counts(dropna=False)
```

Hagamos lo mismo con la columna  **observacion en los arbolers** "above_ground_sighter_measurement", y con las diferentes columnas para ir eliminando valores nulos **NaN**

```python
df_ardilla['above_ground_sighter_measurement'].value_counts(dropna=False)
```

```python
df_ardilla['above_ground_sighter_measurement']=df_ardilla['above_ground_sighter_measurement'].fillna('FALSE')
df_ardilla['above_ground_sighter_measurement'].value_counts(dropna=False)
```

En este caso con este Array convertimos el color **Nan** en un **Distint**

```python
df_ardilla['primary_fur_color'].value_counts(dropna=False)
```

```python
df_ardilla['primary_fur_color']=df_ardilla['primary_fur_color'].fillna('Distint')
df_ardilla['primary_fur_color'].value_counts(dropna = False)
```
